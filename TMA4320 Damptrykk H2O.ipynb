<b> Innledning

I dette prosjektet skal vi finne damptrykk-kurven til $H_2O$ ved hjelp av van der Waals ligning og Clapeyrons ligning. Damptrykk-kurven beskriver trykket ved likevekten til gass- og væske-fasen til et stoff, og går fra trippelpunktet til kritisk punkt til stoffet. I første oppgave ser vi hovedsakelig på van der Waals ligning og løser denne med Newtons metode. I andre oppgave vil vi fokusere på Clapeyrons ligning som kan løses med numerisk integrasjon, i vårt tilfelle med Simpsons metode. I begge oppgavene vil vi sammenligne med eksperimentelle verdier og se hvor godt ligningene og de numeriske metodene klarer å beskrive damptrykk-kurven.

<b> Teori/motivasjon

For å finne damptrykk-kurven til $H_2O$ kan man bruke at stigningstallet til damptrykket er gitt ved

$\frac{dp}{dT}=\frac{\Delta S}{\Delta V} \tag{1}$

der $\Delta S = S_g-S_v$ er entropien og $\Delta V = V_g-V_v$ er volumet når en viss stoffmengde omdannes fra væske til gass, ved trykk $p$ og temperatur $T$ . Dette er Clapeyrons ligning som kan integreres for å finne $p(T)$ slik:

$p(T) = p_t +\int_{T_t}^{T}\frac{L(T)dT}{T\Delta V(T)} \tag{2}$

Her ble det brukt at $\Delta S = \frac{L(T)}{T}$, der $L$ er fordampningsvarmen. Siden damptrykkurven til vann går fra trippelpunktet til kritisk punkt, brukes trippelpunketet $T_t$ som referanse.
For å finne $p(T)$ må man først finne ligninger som beskriver $L(T)$, samt $\Delta V(T)$ for så å kunne integrere ligningen numerisk. For å finne en analytisk løsning for damptrykkurven må det gjøres et par antakelser. For det første antas det at fordampningsvarmen $L$ er konstant i et lite temperaturområde rundt et valgt punkt $(p_0,T_0)$. Hvis man er langt nok unna kritisk punkt vil gassvolumet være mye større enn væskevolumet. Sammen med å anta ideell gass gir dette $\Delta V\approx V_g = \frac{nRT}{p} $. Hvis man nå integrerer clapeyrons ligning får man

$p(T) = p_0exp[\frac{L}{R}(\frac{1}{T_0}-\frac{1}{T})] \tag{3}$

En tilnærming til damptrykkurven kan finnes med van der Waals ligning, som er en forbedret versjon av den ideelle gassloven. Ideell gasslov er gitt ved:

$p = \frac{nRT}{V} \tag{4}$

der $n$ er antall mol, $R$ er universell gasskonstant, $V$ er volumet og $T$ er temperatur.
Van der Waals ligning er gitt ved

$p = \frac{RT}{V-b}-\frac{a}{V^2} \tag{5}$

dersom det er 1 mol med stoff. Siden 1 mol med $H_2O$ vil studeres gjennom hele prosjektet er det kun denne formen av ligningen som brukes videre. Parametrene $a$ og $b$ justerer respektivt for de svake tiltrekkende kreftene mellom molekyler som gir et redusert trykk og at molekyler ikke er punktpartikler, og dermed har tilgjengelig volum som er mindre enn det totale volumet $V$. $a$ og $b$ er empiriske og fastlegges vanligvis med to av de tre størrelsene $T_c$, $p_c$ og $V_c$ i kritisk punkt. Van der Waals ligning er kun er en forbedring i forhold til ideell gasslov med tanke på å finne damptrykkkurven. Den er altså ikke eksakt.

En av de viktige feilene til van der Waals ligning er at for $T<T_c$ vil den gi ufysiske verdier for volumer mellom $V_v$ og $V_g$. Det burde gi en konstant koeksistenskurve, men det vil i stedet bli en kurve som krysser denne linjen i tre punkt. For å finne trykket til koeksistenskurven, $p_0$, kan Maxwells konstruksjon benyttes. Av de tre punktene  er den med lavest volum $V_v$, den med høyest volum er $V_g$ og punktet i midten er benevnet $V_0$. $V_0$ er plassert slik at arealet mellom van der Waals kurven og likevektskurven mellom $V_v$ og $V_0$ er det samme som arealet mellom likevektskurven og van der Waals kurven mellom $V_0$ og $V_v$. Dette gir dermed to ligninger for $V_v$ og $V_g$ som i dette prosjektet skal løses med Newtons metode. Ligningene er gitt ved

$$\frac{RT}{V_g-b}-\frac{a}{V_g^2}-\frac{RT}{V_v-b} + \frac{a}{V_v^2}  \tag{6}$$

og 

$$\frac{RT}{V_g-V_v}log(\frac{V_g-b}{V_v-b})-\frac{a}{V_gV_v} -\frac{RT}{V_g-b}+\frac{a}{V_g^2} \tag{7}$$

der alle parametrene er nevnt i teorien over (Støvneng, et al. 2023. s.5).



## <center>  Oppgave 1: van der Waals tilstandsligning og damptrykk

<b> 1a

Van der Waals ligning er gitt ved ligning (4). Parametrene $a$ og $b$ er empiriske og kan bestemmes med denne relasjonen for kritisk temperatur og trykk:

$$(\frac{\partial p}{\partial V})_c = (\frac{\partial^2 p}{\partial V^2})_c = 0$$

Ut ifra disse realsjonene kommer man frem til

$$V_c = 3b, T_c = \frac{8a}{27Rb}, p_c = \frac{a}{27b^2}$$

der $V_c$, $T_c$ og $p_c$ er henholdsvis de kritisk verdiene for volum, trykk og temperatur. $R$ er den universelle gasskonstanten.

Ved å bruke ligningene for $p_c$ og $T_c$ kan en finne utrykk for $a$ og $b$:

$$a = \frac{27R^2T_c^2}{64p_c}, b = \frac{RT_c}{8p_c}$$

Det kritiske molare volumet med disse ligningene blir $V_c = \frac{3RT_c}{8p_c}$, som med verdier for $T_c$ og $p_c$ gir $V_c = 9.144 * 10^{-5} m^3 = 91.44 ml$. Eksperimentell verdi for $V_c = 55.948 ml$. Det er dermed klart at Van der Waals tilstandsligning ikke gir samsvar med eksperimentell verdi, og ikke kan brukes for kvantitative resultater (Støvneng, et al. 2023. s. 4).

Trykket kan plottes som funksjon av volumet med konstant temperatur ved hjelp av van der Waals tilstandsligning. Under brukes temperaturen $T = T_c$. 

```python
# Importerer nødvendige biblioteker
import numpy as np
import pandas as pd
import scipy.optimize
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [10, 6]
```

```python
# Oppretter nødvendige konstanter
R = 8.314       
Tc = 647.096
pc = 22.064*10**6
a = 27*R**2*Tc**2/(64*pc)
b = R*Tc/(8*pc)
```

```python
def p(V, T):
    '''
    Tar inn et volum og regner ut trykket til 1 mol H2O ved kritisk temperatur
    ---------------------
    Input:
        V: volum til fluidet (m^3)
        T: temperatur (K) 
    Return:
        P: trykk til fluidet (bar)
    '''

    #regner ut trykk med Van der Waals ligning
    P = (8.314*T/(V-b) - a/V**2)/(10**5)
    return P


#plotter trykk vs volum ved 647 K
v = np.linspace(75, 300, 225)*10**-6
plt.title('1 mol $H_2O$, van der Waals isoterm 647 K')
plt.xlabel('Volum (ml)')
plt.ylabel('Trykk (bar)')
plt.plot(v*10**6, p(v, 647.096))
plt.show()
```

Dette gir mening fordi ved kritisk temperatur vil trykkurven ikke passere gjennom damptrykkurven, men tangere den ved kritisk punkt. Det betyr og at det ikke er noen ufysiske punkter i denne grafen. 

Dersom Van der Waals ligning brukes til å finne en isoterm i et pV diagram, så vil den gi ufysiske verdier med $T$ under $T_c$. For disse temperaturene vil trykket mellom $V_v$ og $V_g$  være konstant, siden dette beskriver faseovergangen mellom vsæke og gass (Støvneng, et al. 2023. s.4).

<b> 1c

Newtons metode skal nå implementeres for én variabel og brukes til å løse en testligning. Løsningen på testligningen er $\frac{2}{ln(1+\sqrt(2))}\approx 2.692$. Newtons metode for én variabel løser en ligning på formen $f(x) = 0$ for $x$. Den kan formuleres slik:

$$x_{n+1} = x_n + \frac{f(x_n)}{f'(x_n)}$$

der $x_n$ er startverdi for en iterasjon. Rekken {$x_n$} vil konvergere mot roten til $f(x)$ dersom startverdien $x_0$ er nærme nok til roten.

```python
def Newton1(x_0, func, funcD, tollDiff):
    '''
    Løser ligning numerisk med Newtons metode, og returnerer løsning 
    ---------------------
    Input:
        x_0: Startverdien for iterasjon (K)
        func: Funksjonen som brukes for iterasjon
        funcD: Derivert av func
        tollDiff: toleransen for hvor liten forskjell en iterasjon skal gi før løkken brytes
    Return:
        x: array med itererte x verdier, siste verdi er numerisk løsning (K)
    '''
    
    # Oppretter liste med startverdi for x
    x = []
    x.append(x_0)
    # Setter maks antall iterasjoner
    for i in range(20):
        
        # Sjekker for å kjøre minst 3 ganger
        if i > 2:
            
            # Sjekker for stor nok endring mellom forrige 2 iterasjoner til å iterere igjen
            if (abs(x[-1] - x[-2])) > tollDiff:
                
                # Regner ut neste verdi
                x.append(x[-1]-func(x[-1])/funcD(x[-1]))
            else:
                break
        else:
            # Regner ut neste verdi
            x.append(x[-1]-func(x[-1])/funcD(x[-1]))
            
    return np.array(x)


def func(Tc):
    '''
    Testfunksjon
    ---------------------
    Input:
        Tc: kritist temperatur
    Return:
        verdien av funksjonen i Tc
    '''
    try:
        return np.sinh(2/Tc)**2-1
    except:
        return 0

def funcD(Tc):
    '''
    Derivert av testfunksjonen
    ---------------------
    Input:
        Tc: kritist temperatur
    Return:
        verdien av den deriverte til funksjonen i Tc
    '''
    try:
        return 2*np.sinh(2/Tc)*np.cosh(2/Tc)*(-2/Tc**2)
    except:
        return 1




print('Konvergens for startverdier mellom -5 og 5, (ikke 0 fordi dette gir division by zero error): \n')

newtonTest = Newton1(1, func, funcD, 0.0000000001)

newtonTestList = []
for i in range(-5,6):
    if i != 0:
        newtonTestList.append(Newton1(i, func, funcD, 0.0001)[-1])
print(newtonTestList)
```

Newtons metode er veldig praktisk for å løse en ligning, og konvergerer raskt mot en nøyaktig tilnærming, gitt en god start verdi. Ikke alle startverdier fører til at metoden konvergerer mot riktig svar. Ut ifra svarene for startverdiene mellom -5 og 5 ses det at man får riktig svar for startverdier [-4, 1, 2, 3]. Startverdien påvirker numerikken i relativt liten grad hvis startverdien er god nok til å faktisk konvergere, fordi Newtons metode konvergerer raskt. Om startverdien derimot ikke er nærme nok, er det ikke sikkert at metoden konvergerer, og dette kan føre til overflow, og division by zero feil. 

<b> 1d

Fordi den analytiske løsningen til testfunksjonen er kjent, kan feilen beregnes, og dermed er det mulig å finne konvergensgraden numerisk. Dette gjøres ved hjelp av følgen $p_i$, som konvergerer mot konvergensgraden $q$. Rekken $p_i$ er definert som 

$$p_i = \frac{log(e_i/e_{i-1})}{log(e_{i-1}/e_{i-2})} $$

der $e_i$ er feilen for $i$-te steg
(Støvneng, et al. 2023. s. 8).

```python
def error(analVerdi, numLos):
    '''
    Tar inn en analytisk verdi, og en array, og regner ut feilen i de numeriske verdiene
    ---------------------
    Input:
        analVerdi: analytisk løsning 
        numLos: numeriske løsninger
    Return:
        e: array med feiler 
    '''
    # Oppretter tom array
    e = np.zeros(len(numLos))
    
    # Itererer gjennom numeriske verdier, og legger absolutt feil til arrayet
    for i in range(len(numLos)):
        e[i] = abs(analVerdi-numLos[i])
    return e


def grad(e):
    '''
    Regner ut forholdet mellom logaritmen til endringen in feilen i to følgende steg for å bestemme konvergensgraden til Newtons metode
    ---------------------
    Input:
        e: array med feiler
    Return:
        p: forholdet mellom logaritmen til endringen i feil mellom 2 steg og de forrige 2 stegene
    '''
    # Oppretter tom array
    p = np.zeros(len(e)-2)
    
    # Itererer og regner ut p verdier
    for i in range(2,len(e)):
        p[i-2] = (np.log(e[i]/e[i-1]))/(np.log(e[i-1]/e[i-2]))
    return p


# Regner ut analytisk verdi, og plotter beregnet feil
TcA = 2/(np.log(1+np.sqrt(2)))
print('Analytisk verdi: ', TcA)
plt.title('Feil med Newtons metode')
plt.plot(error(TcA, newtonTest))
plt.yscale("log")
plt.ylabel('Feil (K)')
plt.xlabel('i')
plt.show()


# Plotter p-rekken
plt.plot(grad(error(TcA, newtonTest)))
plt.title('$p_i$ verdier')
plt.ylabel('$p_i$')
plt.xlabel('i')
plt.show()
```

Fordi grafen til feilen minker raskere enn en lineær funksjon, er det tydelig at Newtons metode konvergerer superlineært. For å bestemme nøyaktig konvergensgrad må p-rekken studeres.

Det er kjent at p-rekken vil konvergere mot $q$ for en gitt metode. Ut ifra grafen ser man at følgen $p_i$ konvergerer mot 2, som viser at Newtons metode har konvergensgrad lik 2.

<b> 1e

For å finne damptrykkurven numerisk må ligningsettet gitt ligningene (6) og (7) løses for $V_v$ og $V_g$. I denne oppgaven blir Newtons metode med to variabler implementert for å finne molart gassvolum og molart væskevolum på damptrykkkurven.

```python
def Newton2(T, x_0, y_0, F, G, Fdx, Fdy, Gdx, Gdy, tollDiff):
    '''
    Løser et ligningsett med 2 ukjente med newtons metode
    ---------------------
    Input:
        T: Temperaturen ligningsettet løses for
        x_0: Startverdien til x
        y_0; Startverdien til y
        F: Den første funksjonen i systemet
        G: Den andre funksjonen i systemet
        Fdx: F derivert med hensyn på x
        Fdy: F derivert med hensyn på y
        Gdx: G derivert med hensyn på x
        Gdy: G derivert med hensyn på y
        tollDiff: toleransen for hvor liten forskjell en iterasjon skal gi før løkken brytes
    Return:
        X: Alle x verdier funnet gjennom iterasjon, den siste verdien er numerisk løsning
        Y: Alle y verdier funnet gjennom iterasjon, den siste verdien er numerisk løsning
        
    '''
    #oppretter startverdier og lister for x og y verdier
    X = [x_0]
    Y = [y_0]
    #itererer ikke mer en 20 ganger
    for i in range(20):
        #itererer minst 3 ganger
        if i > 2:
            #sjekker om forbedringen for hver iterasjon er over toleransen
            if X[-1]**2 + Y[-1]**2 > tollDiff**2:
                #lager array med verdier for Jacobi matrisen, og funksjonen
                J = np.array([[Fdx(X[i], Y[i], T), Fdy(X[i], Y[i], T)],[Gdx(X[i], Y[i], T), Gdy(X[i], Y[i], T)]])
                f = np.array([F(X[i], Y[i], T), G(X[i], Y[i], T)])
                
                #løser ligningsettet for delta x, og finner nye verdier for X og Y
                x = np.linalg.solve(J, -f)
                X.append(X[-1] + x[0])
                Y.append(Y[-1] + x[1])
            #Hvis endringen mellom 2 iterasjoner er mindre en tollDiff så stopper løkken
            else:
                break
        else:
            #regner ut minst 3 steg
            J = np.array([[Fdx(X[i], Y[i], T), Fdy(X[i], Y[i], T)],[Gdx(X[i], Y[i], T), Gdy(X[i], Y[i], T)]])
            f = np.array([F(X[i], Y[i], T), G(X[i], Y[i], T)])
    
            x = np.linalg.solve(J, -f)
            X.append(X[-1] + x[0])
            Y.append(Y[-1] + x[1])
    
    return X, Y
```

```python

# Vg = x, Vv = y
def F(Vg, Vv, T):
    '''
    En av funksjonene i ligningsettet
    ---------------------
    Input:
        Vg: Gassvolum
        Vv: Væskevolum
        T: Temperatur
    Return:
        Verdien til en av sidene til ligningsettet meg gitte verdier
    '''

    return R*T/(Vg-b)-a/Vg**2-R*T/(Vv-b)+a/Vv**2

def Fdx(Vg, Vv, T):
    '''
    F derivert med hensyn på Vg
    ---------------------
    Input:
        Vg: Gassvolum
        Vv: Væskevolum
        T: Temperatur
    Return:
        Verdien til F derivert med hensyn på Vg meg gitte verdier
    '''

    return -R*T/(Vg-b)**2+2*a/Vg**3

def Fdy(Vg, Vv, T):
    '''
    F derivert med hensyn på Vv
    ---------------------
    Input:
        Vg: Gassvolum
        Vv: Væskevolum
        T: Temperatur
    Return:
        Verdien til F derivert med hensyn på Vv meg gitte verdier
    '''
    
    return R*T/(Vv-b)**2-2*a/Vv**3

def G(Vg, Vv, T):
    '''
    En av funksjonene i ligningsettet
    ---------------------
    Input:
        Vg: Gassvolum
        Vv: Væskevolum
        T: Temperatur
    Return:
        Verdien til en av sidene til ligningsettet meg gitte verdier
    '''

    return R*T/(Vg-Vv)*np.log((Vg-b)/(Vv-b)) - a/(Vv*Vg)-R*T/(Vg-b)+a/Vg**2

def Gdx(Vg, Vv, T):
    '''
    G derivert med hensyn på Vg
    ---------------------
    Input:
        Vg: Gassvolum
        Vv: Væskevolum
        T: Temperatur
    Return:
        Verdien til G derivert med hensyn på Vg meg gitte verdier
    '''

    return -R*T*np.log((Vg-b)/(Vv-b))/(Vg-Vv)**2 + R*T/((Vg-b)*(Vg-Vv)) + a/(Vg**2*Vv) + R*T/(Vg-b)**2 - 2*a/Vg**3

def Gdy(Vg, Vv, T):
    '''
    G derivert med hensyn på Vv
    ---------------------
    Input:
        Vg: Gassvolum
        Vv: Væskevolum
        T: Temperatur
    Return:
        Verdien til G derivert med hensyn på Vv meg gitte verdier
    '''
   
    return R*T*np.log((Vg-b)/(Vv-b))/(Vg-Vv)**2 - R*T/((Vg-Vv)*(Vv-b)) + a/(Vg*Vv**2)
```

```python
# Opretter array for temperaturer og væske- og gass-volum
T = np.arange(274, 648, 1)
Vg = np.zeros(len(T))
Vv = np.zeros(len(T))
VgTrue = np.zeros(len(T))
VvTrue = np.zeros(len(T))

# Oppretter startverdier
Vg_0 =     12.6*10**-3
Vv_0 =     35.7*10**-6

# Itererer gjennom T verdier og regner ut Vv og Vg
for i in range(len(T)):
    
    # Hvis det ikke er første iterasjon endres startverdi for Newtons metode til forrige svar
    if i > 0:
        Vg_0 = Vg[i-1]
        Vv_0 = Vv[i-1]
    c, d = Newton2(T[i], Vg_0, Vv_0, F, G, Fdx, Fdy, Gdx, Gdy, 0.0001)
    Vg[i] = c[-1]
    Vv[i] = d[-1]

    
# Plotter Vv og Vv som funksjon av T
plt.plot(T, Vg*1000, label = "Gass")
plt.plot(T, Vv*1000, label = "Væske")
plt.legend()
plt.xlabel('temperatur (Kelvin)')
plt.ylabel('molart volum ($l/mol$)')
plt.title("Vg og Vv, Newtons metode")
plt.show()
```

Ut ifra plottet for $V_g$ og $V_v$ , kommer det fram at når temperaturen går mot $T_c$ så vil forskjellen mellom $V_g$ og $V_v$ gå mot null. I kritisk punkt vil det ikke lenger være forskjell mellom væske- og gass-fasen, som forklarer hvorfor $\Delta V\rightarrow 0$.

Siden datasettene for $L(T)$, $V_g(T)$ og $V_v(T)$ kun har et begrenset antall verdier for $T$, er det vanskelig å integrere van der Waals ligning. Derfor er det ønskelig å finne tilnærmede funskjoner som beskriver $L$, $V_g$ og $V_v$ og som kan deriveres slik at disse kan brukes i Simpsons metode. 

```python
#---------------------------- Latent varme ----------------------------------------------

# Finner funskjon som gir god tilpasning til data for latent varme, L:
def func_l(x, a, b, c, d, e):                        # Denne skal optimaliseres
    return a+b*x+c*x**2+abs(647.096-x)**d+e*x**3         # Tc = 647.096K

l_opt, l_cov = scipy.optimize.curve_fit(func_l, temp_L, latent_varme)



# Plot latent varme
plt.plot(temp_L, latent_varme)                        # Eksperimentell måleserie
plt.plot(temp_L, func_l(temp_L, *l_opt), "r--")       # Curve fit
plt.title("Latent varme (L) H2O")
plt.legend(["eksperimentell måleserie", "L(t) = a+bt+ct^2+|Tc-t|^d+et^3"])
plt.xlabel("Temperartur (K)")
plt.ylabel("L ( J/mol )")
plt.text(450, 40000, f"a={l_opt[0].round(2)},b={l_opt[1].round(2)},c={l_opt[2].round(2)},d={l_opt[3].round(2)}, e={l_opt[3].round(2)}")
plt.show()


#------------------------------ Vv ---------------------------------------------------

# Finner funskjon som gir god tilpasning til data for væskevolumet, Vv:
def func_Vv(x, a,b, c):
    return a*x**2+b*x+c                

Vv_opt, Vv_cov = scipy.optimize.curve_fit(func_Vv, temp_vann, np.log(Vv_eksp))  # Bruker log for å tydeligere se hvor god tilpasningen er

# Plot Vv
plt.plot(temp_vann, Vv_eksp)
plt.plot(temp_vann, np.exp(func_Vv(temp_vann, *Vv_opt)), "r--")  
plt.title("Vv")
plt.xlabel("Temperartur (K)")
plt.ylabel("Volum (m^3/mol)")
plt.legend(["eksperimentell måleserie", "Vv(t) = exp{at^2+bt+c}"])
plt.text(265, 5e-5, f"a={Vv_opt[0].round(7)}, b={Vv_opt[1].round(4)}, c={Vv_opt[2].round(7)}")
plt.show()


#------------------------------ Vg ---------------------------------------------------

# Finner funskjon som gir god tilpasning til data for gassvolumet, Vg:
def func_Vg(x, a, b,c,d,e):
    return a*x**4+b*x**3+c*x**2+d*x+e                   

Vg_opt, Vg_cov = scipy.optimize.curve_fit(func_Vg, temp_gass, np.log(Vg_eksp))  # Bruker log for å tydeligere se hvor god tilpasningen er


# Plot Vg
plt.plot(temp_gass, (Vg_eksp))
plt.plot(temp_gass, np.exp(func_Vg(temp_gass, *Vg_opt)), "r--")
plt.title("Vg")
plt.xlabel("Temperartur (K)")
plt.ylabel("Volum (m^3/mol)")
plt.legend(["eksperimentell måleserie", "Vg(t) = exp{at^4+bt^3+ct^2+dt+e}"])
plt.text(470, 3, f"a={Vg_opt[0].round(10)}, b={Vg_opt[1].round(7)}, c={Vg_opt[2].round(4)}, d = {Vg_opt[3].round(2)}, e = {Vg_opt[4].round(2)}")
plt.show()
```

For å finne passende funksjoner brukes scipy.optimize.curve_fit(). Ved å plotte den tilnærmede funskjonen samt de eksperimentelle verdiene kan man prøve seg fram til man finner en god funksjon. Ved hjelp ifra hintet til oppg 2b) fra oppgavebeskrivelsen ble det funnet en god tilpasning til latent varme. Først fant vi andre tilnærminger til $V_v$ og $V_g$ enn i koden over, men dette skapte problemer med å få en god tilpasning når Simpsons metode ble tatt i bruk. <b> 2c

For å finne et numerisk anslag for $p(T)$ på det lukkede intervallet $[274, 647]$ $K$ vil integranden til ligning $(2)$ integreres ved hjelp av Simpsons metode. Dette er en numerisk metode som tar utgangspunkt i at en graf $f$, kan tilnærmes med segmenter av parabler (Adams & Essex, 2021). I denne deloppgaven vil den numeriske integrasjonen gjøres fra $T_t$ til $T$ for alle heltallstemperaturer i det angitte intervallet, og i integranden vil utrykkene for hensholdsvis latent varme, væskevolum og gassvolum som ble fremstillt ved kurvetilpassning i oppgave 2b) benyttes. Det numeriske anslaget for $p(T)$ vil så plottes sammen med den eksperimentelle damptrykk-kurven for å kunne vurdere hvor god den numeriske tilnærmingen er.    

For å finne $p(T)$ integreres Clapeyrons ligning numerisk slik at vi får ligning (2).

Den numeriske integrasjonen gjøres med Simpsons metode, som tilnærmer integralet med følgende uttrykk 

$$ \int_{a}^{b} f(x) \,dx\ \ \simeq \frac{h}{3}(y_0 + 4y_1 + 2y_2 + 4y_3 + .. + 2y_{n-2} + 2y_{n-1} + y_n)$$

hvor 
$$h = \frac{(b-a)}{n}, y_i= f(a+ih)$$

og $n$ er et like tall.



```python
def simpson(a, b, f, n): 
    
    """
       Evaluerer integralet av f(x) på intervallet [a,b] med Simpsons metode 
       ---------------------------------------------------------
       Input: 
       - a, b = venstre og høyre endepunkt av integrasjons intervall
       - f = integrand
       - n = antall underintervaller (må være et partall)
       
       Output:
       - S = numerisk tilnærmelse av integralet  
    """
    
    
    assert(n%2==0), "n må være et partall!"     # Sjekker at n er et partall 
        
    
    h = (b-a)/n                                 # Steglengde 
    
    discPoints = np.linspace(a,b, n+1)          # n+1 jevnt fordelte diskrete punkter i intervallet [a,b]

        
    # Beregner tilnærmelse til integral 
    S = h/3 * (f(discPoints[0]) + 4*sum(f(discPoints[1:-1:2])) + 2*sum(f(discPoints[2:-1:2])) + f(discPoints[-1]))
 
    return S 
```

For å sikre at implementasjonen av Simpsons metode i funksjonen simpson() er riktig, ble den testet på en rekke testproblemer med kjent løsning. For eksempel ble tilnærmelsen til integralet av cos(x) på intervallet [0, $\pi$/2],  betraktet med n = 12 steg, beregnet til å være 1,00000163, hvilket svarer til et avvik fra reel verdi på kun 1,63e-6. På bakgrunn av disse testene fikk man bekreftet at funksjonen implementerer den numeriske metoden riktig. 

```python
def clapeyron(T): 
    
    """
       Beregner funksjonsverdien av integranden gitt ved ligning (8) for T 
       -------------------------------------------------------------------
       Input: 
       - T = float verdi / array med float verdier 
       
       Output: 
       - funkVerdi = fuksjonsverdien(e) av integranden for T 
    """
    
    
    # Latent varme som funksjon av T 
    lVarme = func_l(T, *l_opt)
    
    # Væskevolum som funksjon av T                # Bruker Curve fits fra oppgave 2b) 
    vVolum = np.e**(func_Vv(T, *Vv_opt))
    
    # Gassvolum som funksjon av T 
    gVolum = np.e**(func_Vg(T, *Vg_opt))
    
    funkVerdi = lVarme / (T * (gVolum - vVolum))  # Funksjonsverdi av integranden 
    
    return funkVerdi
```

```python
tempInt = np.linspace(274, 647, 374)    # Array med heltallstemperaturer i intervallet T_t til T_c 
trykk = np.zeros(len(tempInt))          # Array for å holde på trykk i de diskrete punktene 
             

# Integrerer ligning (8) numerisk fra trippelpunktet T = 274 til kritisk punkt T = 647 
for i in range(0,len(tempInt)): 
    trykk[i] = simpson(tempInt[0], tempInt[i], clapeyron, 4)

    
# Plotter resultater 
plt.figure(dpi = 100)
plt.plot(tempInt, trykk, label = "Simpsons metode")                                     # Tilnærming fremstilt med Simpsons metode 
plt.plot(temp_1f[:20], p_1f[:20]*1e5, "o", label = "Måledata", color = "orangered")     # Måledata, konverterer til enheten Pa 

# Formaterer plott 
plt.title("Numerisk tilnærming til koeksistenskurven $p(T)$ for $H_2O$\n$V_v$, $V_g$ og $L$ tilnærmet med curve fit")
plt.xlabel("Temperatur T, [K]")
plt.ylabel("Trykk p, [Pa]")
plt.grid(ls= "-.")
plt.legend()
```

Figuren ovenfor viser det numeriske anslaget for $p(T)$ fremstilit med Simpsons metode, hvor kurvetilpassningene for henholdsvis $L, V_v$, og $V_g$ er benyttet i integranden. Plottet viser at anslaget samsvarer godt med de eksperimentelle dataene i store deler av temperaturintervallet som betraktes. Den numerisk fremstilite damptrykkurven begynner først å få betydelig avvik når temperaturen begynner å nærme seg området rundt kritisk temperatur $T_c$. I plottet ovenfor er ikke tilnærmingen for trykket i $T$ = $T_c$ inkludert, ettersom denne verdien blir veldig stor som følge av at $V_g$ og $V_v$ praktisk talt blir like store i dette punktet. Dette medfører at man i integranden deler på tilnærmet lik null, slik at man får en meget stor verdi. I området med avvik synlig på plottet, observerer man at den numeriske tilnærmingen overestimerer trykket. Dette avviket kan sees i sammenheng med avvikene i kurvetilpassningen som benyttes i integranden, ettersom disse følgelig vil videreføres. Dersom plottene i 2b) av kurvetilpassningene sammen med de eksperimentelle verdiene betraktes, observerer man at tilnærmingene for gassvolm og latent varme på generelt grunnlag er gode, men at kurvetilpasning for væskevolum gir en dårlig tilnærmelse til de eksperimentelle dataene. I området hvor avviket til kurvetilpasning for væskevolum er størst, overestimeres volumet. For integranden gitt i ligning (2) medfører dette at funksjonsverdien blir mye større enn den reelle verdien ettersom nevneren blir mindre. Dette er en faktor som kan bidra til det avviket som observeres i det numeriske anslaget ettersom estimatet som fremstilles med Simpsons metode er basert på $2^n + 1$ funksjonsverdier av nettopp integranden på det gitte intervallet [a,b]. Man må imidlertid også ta i betraktning at den numeriske metoden kun gir et anslag, og at det derfor alltid vil være avvik tilstede. I dette tilfellet ville et valg av flere steg i hvert intervall kunne ha minsket feilen direkte knyttet til den numeriske integrasjonen.

<b> 2d

I oppgave 2b) benyttes scipy funksjonaliteten scipy.optimize.curve_fit() til å finne funksjonsuttrykk for henholdsvis latent varme, væskevolum og gassvolum som funksjon av temperatur $T$. Denne funksjonen bruker minste kvadraters metode for ikke-lineære funksjoner for å tilpasse en funksjon $f$ til dataene. Det finnes imidlertid flere metoder for å interpolere data, og i denne deloppgaven vil kubiske spliner benyttes til å finne tilnærmelser til $V_v$, $V_g$ og $L$ i temperaturintervallet $T \in [274, T_c]$. En kubisk-spline-funksjon $S(x)$ kan sees på som en funksjon som tilnærmer et datasett ved hjelp av et sett med kubiske polynomer som er "sydd sammen"(Støvneng, et al. 2023. s.11). For å kunne sammenligne hvor godt de ulike metodene for interpolasjon greier å representere de eksperimentelle dataene, vil disse, samt tilpassningsfunksjonene fra oppgave 2b) plottes sammen med de eksperimentelle måledataene. På bakgrunn av disse plottene vil hvilke type interpolasjon som egner seg best for hvert av datasettene vurderes. 

```python
# --------------------------GASSVOLUM--------------------------

intpolGas = CubicSpline(temp_gass, Vg_eksp)   # Interpolerer data med kubiske spliner 

tempGas = np.linspace(274, 647, (647-274)*4)  # Array med jevnt fordelte punkter i temperatur intervallet [290, 520] K  


figure, axis = plt.subplots(1, 3, dpi = 100, figsize=(12,3))
  
# Kubiske spliner 
axis[0].semilogy(tempGas, intpolGas(tempGas), label = "$S(x)$", color = "steelblue")  # Interpolasjon med kubiske spliner
axis[0].semilogy(temp_gass, Vg_eksp, ".", label = "Måledata", color = "orangered")    # Måledata 
axis[0].set_title("Kubiske spliner") 
axis[0].grid(ls="-.")
axis[0].legend()
axis[0].set_ylabel("Gassvolum $V_g$, [$m^3/mol$]")

  
# Sammenligning av interpolasjoner 
axis[1].semilogy(tempGas, intpolGas(tempGas), "-" ,label = "$S(x)$", color = "steelblue")                       # Interpolasjon med kubiske spliner
axis[1].semilogy(tempGas, np.exp(func_Vg(tempGas, *Vg_opt )), "-", label = "Curve fit", color = "darkorange")   # Kurvetilpassning fra 2b) 
axis[1].set_title("Interpolasjon av eksperimentell data\nGassvolum $V_g(T)$")
axis[1].set_xlabel("Temperatur T, [K]")
axis[1].grid(ls="-.")
axis[1].legend()


# Kurvetilpassning 
axis[2].semilogy(tempGas, np.exp(func_Vg(tempGas, *Vg_opt )), "-", label = "Curve fit", color = "darkorange")  # Kurvetilpassning fra 2b) 
axis[2].semilogy(temp_gass, Vg_eksp, ".", label = "Måledata", color = "orangered")                             # Måledata 
axis[2].set_title("Curve fit")
axis[2].grid(ls="-.")
axis[2].legend()
plt.show()
```

Med utgangspunkt i plottene ovenfor kommer det frem at begge metodene for interpolasjon av de ekseperimentelle dataene greier å representere datasettet knyttet til gassvolum i tilfredsstillende grad. I begge tilfellene samsvarer kurvene godt med datapunktene over hele temperaturintervallet, og resultatene fremstilt med de to metodene samsvarer godt med hverandre. Interpolasjonen gjort med kubiske spliner er imidlertid noe bedre å representere datasettet nøyaktig. Dette blir tydelig mot slutten av intervallet. På generelt grunnlag greier begge metodene å representere dataene godt på intervallet, og basert på plottene ovenfor er det ikke en metode som veldig klart peker seg ut som best egnet til å interpolere datasettet knyttet til gassvolum. <b> 2e

I oppgave 2c) benyttes kurvetilpassningene for $V_v$, $V_g$ og $L$ fra oppgave 2b) i uttrykket for $dp$ gitt ved ligning (2). I denne deloppgaven vil disse uttrykkene byttes ut med interpolasjonene fremstilt med kubiske spliner i 2d) når den numeriske tilnærmingen til integralet fremstilles med Simpsons metode. Deretter vil det resulterende anslaget for koeksistenskurven $p(T)$ plottes sammen med måledataene som funksjon av temperatur $T$ for kunne vurdere hvor god tilnærmelsen er.

```python
def clapeyronSpline(T):
    
    """
       Beregner funksjonsverdien av integranden gitt ved ligning (8) for T 
       V_v, V_g og L gitt ved tilnærming fremstilt med interpolasjon med kubiske spliner 
       ---------------------------------------------------------------------------------
       Input: 
       - T = float verdi / array med float verdier 
       
       Output: 
       - funkVerdi = fuksjonsverdien(e) av integranden for T 
    """
    
    # Array med de diskrete datapunktene som benyttes ved integrasjonen 
    temps = np.arange(274, 647 + 0.25, 0.25)

    

    # Interpolasjoner for alle de diskrete datapunktene 
    gassVol = intpolGas(temps)
    væskeVol = intpolVol(temps)
    latentVarme = intpolLat(temps)

    # Beregner funksjonsverdi av integranden i alle de diskrete punktene med interpolasjonene 
    clapInt  = latentVarme / (temps*(gassVol - væskeVol))
    
    
    # Hvis input er en skalar 
    if np.isscalar(T) == True: 
        
        for i in range(len(temps)):
            if temps[i] == T:         # Finner korresponderende index til gitt temperatur for å indexere riktig 
                                      # funksjonsverdi 
                    
                return clapInt[i]     # Returner funksjonsverdi som svarer til gitt temperatur 
    
    
    # Hvis input er et array 
    else: 
        
        funcvals = []                              # Array for å holde på funksjonsverdier 
        
        for k in T:
            for j in range(len(temps)):            # Finner korresonderende indexer til gitte temperaturer 
                                                   # for å indexere riktig funksjonsverdi 
                if temps[j] == k: 
                    funcvals.append(clapInt[j])
                    
    
        return funcvals               # Returnerer funksjonsverdier som svarer til gitte temperaturer 
```

```python
trykkSpline = np.zeros(len(tempInt))    # Array for å holde på trykk 


for i in range(0, len(tempInt)): 
    trykkSpline[i] = simpson(tempInt[0], tempInt[i], clapeyronSpline, 4)


# Plotter resultater 
figure, axis = plt.subplots(1, 2, dpi = 100, figsize=(10,3))
  
# Plot hvor kritisk punkt ikke inkluderes for å få frem at numerisk anslag er godt i store deler av intervallet 

axis[0].plot(tempInt[:-2], trykkSpline[:-2], label = "Simpsons metode")                      # Tilnærming fremstilt med Simpsons metode 
axis[0].plot(temp_1f[:20], p_1f[:20]*1e5, ".", label = "Måledata", color = "orangered")      # Måledata, konverterer til enheten Pa 
axis[0].plot(647,0 , "v", color = "r", label = "Kritisk punkt $T_c$")
axis[0].set_title("Tilnærming uten kritisk punkt $T_c$")
axis[0].grid(ls="-.")
axis[0].legend()
axis[0].set_xlabel("Temperatur T, [K]")
axis[0].set_ylabel("Væskevolum $V_v$, [$m^3/mol$]")

  
# Plot hvor kritisk punkt inkluderes 

axis[1].plot(tempInt, trykkSpline, label = "Simpsons metode")                                # Tilnærming fremstilt med Simpsons metode 
axis[1].plot(temp_1f[:20], p_1f[:20]*1e5, ".", label = "Måledata", color = "orangered")      # Måledata, konverterer til enheten Pa 
axis[1].set_title("Tilnærming frem til kritisk punkt $T_c$")
axis[1].set_xlabel("Temperatur T, [K]")
axis[1].plot(647,0 , "v", color = "r", label = "Kritisk punkt $T_c$")
axis[1].grid(ls="-.")
axis[1].legend()   
```

Plottet til høyre viser også tilnærmelsen helt frem den kritiske temperaturen, hvor man kan se effekten av at væske og gassfasen blir like. Feilen i tilnærmingen til damptrykkurven skyldes mest sannsynlig en kombinasjon av feil knyttet til den numeriske metoden for tilnærmingen av integralet, og avvik ved interpolasjon av datasettene. Når det gjelder feil knyttet til den numeriske metoden, vil det alltid oppstå feil som følge av at Simpsons metode kun fremstiller en tilnærming for funksjoner som ikke er polynomer av tredje grad eller mindre. Denne feilen burde i dette tilfellet imidlertid ikke være for stor ettersom funksjonen som integreres er glatt (har kontinulerlige høyre ordens deriverte), og dermed kan tilnærmes godt med parabeler. For ligning $(3)$ vil man finne en verdi for $L$ som tilnærmer den eksperimentelle måleserien godt, og bestemme referansepunktet $(p_0, T_0)$. I dette plottet vil også de eksperimentelle dataene inkluderes slik at resultatene kan sammenlignes og drøftes på bakgrunn av hvor godt tilnærmingen greier å representere den eksperimentelle damptrykk-kurven.

Clapeyrons ligning under antagelsen om ideell gass og tilnærmelsen om at fordampningsvarmen er konstant i et lite temperaturområde rundt et valgt referansepunkt ($p_0$, $T_0$); 

$$ \frac{dp}{dT} = \frac{Lp}{RT^2}$$

Dersom denne integreres får man den analytiske løsningen for damptrykk-kurven gitt ligning (3).

```python
def pAnalytisk(T):
    
    """
       Beregner analytisk løsning for p(T) gitt ved ligning (9)
       --------------------------------------------------------
       Input: 
       T = array med temperaturverdier 
       
       Output: 
       Analytisk løsning for p(T)

    """
    
    L = 42000    # Latent varme [J/mol], verdi som gir god tilnærming til eksperimentell data 
    pt = 611.73  # Trykk i trippelpunkt, [Pa]
    Tt = 273.16  # Temperatur i trippelpunkt, [K]
    return pt*np.e**(L/R*(1/Tt-1/T)) 
```
